### **Assignment 02: Basic Arithmetic with Two Numbers**

#### **Overview**
Create a Python program that prompts the user to enter two numbers and then performs basic arithmetic operations (addition, subtraction, multiplication, and division) on them. The program should display the results in a clear, user-friendly manner.

#### **Objectives**
- Get comfortable with user input and data type conversion.  
- Practice basic arithmetic operations in Python.  
- Learn how to handle edge cases (e.g., division by zero).  
- Format and present output neatly.

---

### **Detailed Instructions**

1. **Prompt the User for Two Numbers**  
   - Use the `input()` function twice to ask the user for two numeric values.  
   - Convert these input strings to a numeric type (e.g., `float` or `int`).  
   - Example prompt:  
     ```python
     # Pseudocode
     first_num = float(input("Enter the first number: "))
     second_num = float(input("Enter the second number: "))
     ```

2. **Perform Arithmetic Operations**  
   - **Addition**: Add the two numbers and store the result.  
   - **Subtraction**: Subtract the second number from the first and store the result.  
   - **Multiplication**: Multiply the two numbers and store the result.  
   - **Division**: Check if the second number is zero before dividing. If it is zero, handle the scenario by printing a message (e.g., “Cannot divide by zero!”).

3. **Output the Results**  
   - Print each operation's result to the screen.  
   - Clearly label each operation so the user knows which result corresponds to which calculation.  
   - Consider formatting floating-point results to two decimal places (optional).

4. **Test Your Program**  
   - Try several different test cases:
     - **Positive numbers** (e.g., 10 and 3).  
     - **Zero** values (especially as the second input for division).  
     - **Negative numbers** (e.g., -5 and 2).  
     - **Decimal/float inputs** (e.g., 3.5 and 1.25).

5. **Submit Your Work**  
   - Push your code to a GitHub repository named something like `basic-arithmetic-assignment`.  
   - Include a `README.md` that briefly explains what your code does and how to run it.  

---

### **Tips & Tricks**

- **Handling Division by Zero**:  
  Always check if the second number is zero. If it is, handle it gracefully to avoid crashes.

- **User Input Validation** (optional):  
  Use a `try-except` block to catch errors if someone enters invalid data (like letters).

- **Formatting Output**:  
  To make your output more readable, you can use Python’s formatting features:
  ```python
  print(f"Sum: {sum_result:.2f}")
  ```

- **Add Comments**:  
  Make sure to include inline comments explaining each major step or variable in your code.

- **Have Fun**:  
  Consider adding extra operations like exponentiation, modulo (`%`), or floor division (`//`), as long as it doesn’t complicate the basic requirements.
